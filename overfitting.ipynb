polynomial fit (non-linear regression)

$y_{predicted} = w_0X_0 + w_1X_1 + w_2X^2_2 + ... + w_nX^n_n$

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X_and_y = [(0, 1), (1, 3), (2, 2), (3, 5), (4, 7), (5, 8), (6, 8), (7, 9), (8, 10), (9, 12)]
data = pd.DataFrame(X_and_y, columns=['X','y'])
X = data['X']
y = data['y']
pick = np.array([True, True, True, True, True, False, True, True, False, True])

train = data[pick]
X_train = train['X']
y_train = train['y']

validation = data[~pick]
X_val = validation['X']
y_val = validation['y']

L2_train = []
L2_val = []

for n in range(1,5):
  w = np.polyfit(X_train, y_train, n)
  f = np.poly1d(w)
  y_pred_train = f(X_train)
  error_train = y_train - y_pred_train
  L2_train.append(0.5 * (np.mean(error_train**2)))

  y_pred_val = f(X_val)
  error_val = y_val - y_pred_val
  L2_val.append(0.5 * (np.mean(error_val**2)))

print(n)

plt.scatter(X_train, y_train)
plt.scatter(X_val, y_val)
plt.plot(X_train, y_pred_train)

x = np.linspace(0, 9, 100)
# generates 100 points between 0 and 9
plt.plot(x, f(x))

#plt.plot(range(1,6), L2_train, label='training')
#plt.plot(range(1,6), L2_val, label='validation')
#plt.legend()
```
