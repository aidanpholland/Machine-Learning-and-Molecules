```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('delaney-processed.csv')
print(list(data))
# choose x variables from here
```

```python
# our macihne learning code
X = data[['Minimum Degree', 'Molecular Weight', 'Number of H-Bond Donors', 'Number of Rings', 'Number of Rotatable Bonds', 'Polar Surface Area']]
N = len(X)
print(N)
ones = np.ones(N)
Xp = np.c_[ones,X]
y = data['measured log solubility in mols per litre']
y = y.values.reshape(1,-1)

np.random.seed(0)
# makes error same every time
w = 2 * np.random.rand(7) - 1

epochs = 1000000
learning_rate = 0.00003
for epoch in range(epochs):
  y_predicted = w @ Xp.T
  error = y - y_predicted
  L2 = 0.5*np.mean(error**2)

  gradient = -(1/N) * error @ Xp
  w = w - learning_rate * gradient
  
  if epoch%(epochs/10) == 0:
    print(epoch,L2)

print(w)
```

```python
# professional machine learning code
from sklearn.linear_model import LinearRegression

X = data[['Minimum Degree', 'Molecular Weight', 'Number of H-Bond Donors', 'Number of Rings', 'Number of Rotatable Bonds', 'Polar Surface Area']]
y = data['measured log solubility in mols per litre']

reg = LinearRegression().fit(X,y)
print(reg.coef_)
print(reg.intercept_)
# have to print intercept as well because bias is not included as x0
y_predicted = reg.predict(X)
error = y - y_predicted
L2 = 0.5 * np.mean(error ** 2)
print(L2)
```
